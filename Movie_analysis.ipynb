Classifying movies as bad,average and good

```python
def rating(rating):
  if rating>7.0:
    return 'good'
  elif rating>=6.0:
    return 'average'
  else:
    return 'bad'
```

```python
data['class']=data['Rating'].apply(rating)
data[['Title','class']]
```

```python
good_movies = data[data['class']=='good']
percentage_good_movies=(len(good_movies)/len(data))*100
percentage_good_movies
```

```python
average_movies = data[data['class']=='average']
percentage_avg_movies=(len(average_movies)/len(data))*100
```

```python
percentage_avg_movies
```

```python
bad_movies = data[data['class']=='bad']
bad_movies
```

```python
count = data['class'].value_counts()
```

```python
plt.pie(count,labels=['good','bad','avg'])
plt.show()
```

```python
data[data['Genre'].str.contains('action',case=False)]
```

```python
list1=[]

for value in data['Genre']:
  list1.append(value.split(','))

list1
```

```python
one_d =[]

for item in list1:
  for item1 in item:
    one_d.append(item1)
```

```python
one_d
```

```python
uni_list=[]

for item in one_d:
  if item not in uni_list:
    uni_list.append(item)
```

```python
uni_list #types of genres
```

```python
len(uni_list)
```

```python
cnt = Counter(one_d)
dictionary = dict(cnt)
dictionary
```

```python
labels=[]
val=[]

for x,y in dictionary.items():
  labels.append(x)
  val.append(y)
```

```python
plt.pie(val,labels=labels)
plt.legend(labels,bbox_to_anchor=(1.05,1))
plt.show()
```
